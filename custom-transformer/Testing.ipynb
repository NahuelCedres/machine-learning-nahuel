```python
from high import Transformer
import torch 
```

```python
batch_size = 32
sequence_length = 100
dec_seq_length = 100
dmodel = 512
num_head = 8
num_layers = 6
dp = 0.1
device = 'cuda'
src_vocab_size = 5000
tgt_vocab_size = 5000
```

```python
model = Transformer(dmodel, 
                    src_vocab_size, 
                    tgt_vocab_size, 
                 sequence_length, 
                 dec_seq_length, 
                    num_head, 
                    num_layers, 
                    dp, 
                    device)
```

```python
# Generar tensores de entrada y target con el rango correcto
input_tensor = torch.randint(0, dmodel, (32, sequence_length)).to(device)
target_tensor = torch.randint(0, dmodel, (32, dec_seq_length)).to(device)

# Ejecutar el modelo con las correcciones
enc_out= model(input_tensor, target_tensor)
enc_out.shape
```
